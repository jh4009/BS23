```python
img_url = "https://brainstation-23.com/wp-content/uploads/2024/08/bs23_logo-250x40.png"

img_data = requests.get(img_url).content
with open("downloaded_image.jpg", "wb") as f:
    f.write(img_data)

print("Image downloaded successfully!")
```

```python
for i, img in enumerate(img1):
    img_url = img.get('src')
    if img_url and img_url.startswith("http"):
        img_data = requests.get(img_url).content
        with open(f"bs23_images/image_{i}.jpg", 'wb') as f:
            f.write(img_data)
        print(f"Downloaded: image_{i}.jpg")
```

```python
import requests

API_URL = "http://172.16.227.1:8000/upload/"
file_path = "downloaded_image.jpg"

# Open the file in binary mode
with open(file_path, "rb") as file:
    files = {"file": (file_path, file, "image/jpeg")}  # Adjust MIME type if needed
    
    try:
        response = requests.post(API_URL, files=files, timeout=10)  # Added timeout
        response.raise_for_status()  # Raises HTTP errors (e.g., 400, 500)
        
        # Handle non-JSON responses gracefully
        try:
            print("Upload successful:", response.json())
        except requests.exceptions.JSONDecodeError:
            print("Upload successful, but response is not JSON:", response.text)

    except requests.exceptions.RequestException as e:
        print(f"Error uploading file: {e}")
```
